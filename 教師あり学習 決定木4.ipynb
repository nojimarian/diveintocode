```python
import numpy as np
from sklearn.datasets import load_boston
boston = load_boston()
import pandas as pd
X = pd.DataFrame(boston.data[:100,:], columns = boston.feature_names)
y = boston.target[:100]
```

```python
x = X['LSTAT'].values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.scatter(x,y)
```

```python
#モデルの構築
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(max_depth=3)
tree.fit(x[:,None],y)
```

```python
#回帰直線を図示するのに変数を並び替え
sort_idx = x[:,None].flatten().argsort()
#argsort() 並び替えた結果の配列のインデックスを返す
#flatten() ネストされた配列を1次元配列にフラット化する
```

```python
#プロット
plt.figure(figsize = (10,7))
plt.scatter(x[sort_idx], y[sort_idx], c = 'blue', label = 'Training points')
plt.plot(x[sort_idx], tree.predict(x[sort_idx].reshape(100,1)),color = 'red', label = 'depth=3')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.legend(loc = 'upper right')
plt.show()
```

```python
#課題
from sklearn.tree import DecisionTreeRegressor
tree_1 = DecisionTreeRegressor(max_depth = 1)
tree_3 = DecisionTreeRegressor(max_depth = 3)
tree_5 = DecisionTreeRegressor(max_depth = 5)
tree_7 = DecisionTreeRegressor(max_depth = 7)
tree_10 = DecisionTreeRegressor(max_depth = 10)
```
